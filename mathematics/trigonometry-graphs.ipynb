```python
import math
import numpy as np
import matplotlib.pyplot as plt

angles = np.linspace(-np.pi/20,np.pi/20,1000)
angles_deg= np.degrees(angles)

sine = np.sin(angles_deg)
cose = np.cos(angles_deg)
plt.figure(figsize=(10,6))
plt.plot(angles_deg, sine)
plt.plot(angles_deg, cose)
plt.grid(True)
plt.show()
```
